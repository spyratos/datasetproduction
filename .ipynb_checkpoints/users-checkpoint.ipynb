```python
import numpy as np
import pandas as pd
from math import *
import random
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
np.random.seed(3)
```

```python
state=np.random.get_state()
```

```python
df = pd.read_csv('data/usernames1.csv',names=['user_id','username'])
del df['user_id']
```

```python
user=df.sample(10000,random_state=7)
user.reset_index(drop=True,inplace=True)
```

```python
np.random.set_state(state)
user['age'] = (np.random.beta(2,9,len(user))*180)+10
user['age'] = user['age'].astype(int)
sns.set(color_codes=True)
sns.distplot(user['age'])
plt.show()
```

```python
#user['age'].value_counts().sort_index()
```

```python
#create gender with probabilities
np.random.set_state(state)
user['gender'] = np.random.choice(['male','female','na'],len(user),p=[0.45,0.45,0.1])
```

```python
#print user['gender'].value_counts()
```

```python
#create fav genre with probabilities
np.random.set_state(state)
user['favorite_genre'] = np.random.choice(['sci-fi','drama','western','action','comedy','horror','romance'
                                ,'adventure','animation','LORD OF THE RINGS',''],
                               len(user),p=[0.107,0.107,0.107,0.107,0.107,0.107,0.107,0.107,0.107,0.001,0.036])
```

```python
#print user['favorite_genre'].value_counts()
```

```python
np.random.set_state(state)
user['votes'] = (np.random.beta(1.8,1.9,len(user))*300)
user['votes']=user['votes'].astype(int)
sns.set(color_codes=True)
sns.distplot(user['votes'])
plt.show()
```

```python
daf = pd.read_csv('data/movie_names2.csv',names=['movie_id','title'])
del daf['movie_id']
```

```python
movies=daf.sample(300000,random_state=10)
movies.reset_index(drop=True,inplace=True)
```

```python
np.random.set_state(state)
movies['main_genre'] = np.random.choice(['sci-fi','drama','western','action','comedy','horror','romance'
                                ,'adventure','animation','LORD OF THE RINGS',''],
                               len(movies),p=[0.107,0.107,0.107,0.107,0.107,0.107,0.107,0.107,0.107,0.001,0.036])
#print movies['main_genre'].value_counts()
```

```python
np.random.set_state(state)
movies['year'] = (np.random.beta(5.8,2.9,len(movies))*250)+1800
movies['year']=movies['year'].astype(int)
sns.set(color_codes=True)
sns.distplot(movies['year'])
plt.show()
```

```python
%%time
mscores = np.zeros(len(movies))
counts = np.zeros(len(movies))
avgscores = np.zeros(len(movies))
for i in range(len(reviews)):
    mscores[reviews['movie_id'].iloc[i]] += reviews['score'].iloc[i]
    counts[reviews['movie_id'].iloc[i]] += 1
for i in range(len(mscores)):
    if counts[i]!=0:
        avgscores[i] =mscores[i]/counts[i] 
movies['avg_score']=avgscores    
```
